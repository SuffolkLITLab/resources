```python
import pandas as pd
import re
```

```python
counties_df = pd.read_csv("county_info.csv")
villages_df = pd.read_csv("village_info.csv")
display(counties_df)
display(villages_df)
```

```python
def county(text):
    output = None
    global counties_df
    global villages_df
    for i,row in counties_df.iterrows():
        if re.sub("[^a-zA-Z]","",row["Municipality"].lower()) in re.sub("[^a-zA-Z]","",text.lower()):
            output = row["County"]
    if output==None:
        for i,row in villages_df.iterrows():
            if re.sub("[^a-zA-Z]","",row["Village Name"].lower()) in re.sub("[^a-zA-Z]","",text.lower()):
                output = row["County"]        
    return output
```
